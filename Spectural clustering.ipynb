```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import SpectralClustering
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import SpectralClustering 
clustering=SpectralClustering(n_clusters=5,  random_state=None, n_init=10, gamma=1.0, affinity='rbf', n_neighbors=10, eigen_tol=0.0, assign_labels='discretize', degree=3, coef0=1, kernel_params=None, n_jobs=None)
y_mean=clustering.fit_predict(x)
```

```python
y_mean
```

```python
supervised=dataset
```

```python
supervised['Cluster_group']=y_mean
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```
